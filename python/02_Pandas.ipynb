```python
# 시리즈 , dataframe 인덱싱/ 슬라이싱
```

```python
population = pd.Series([9602000,3344000,1488000,2419000], index = ['서울','부산','광주','대구'])
population.index.name = '도시'
population.name = '2020 인구수'
```

```python
population
```

```python
# 인덱싱 방법 -- 1. 인덱스 번호로 접근
population[0]
population[2]
```

```python
# 인덱싱 방법 - 2. 인덱스 이름으로 접근
population['부산']
```

```python
# 여러개의 값을 인덱싱
# 인덱스 번호로 접근 -- 서울, 대구
population[[0,3]]
```

```python
# 2. 인덱스 이름으로 접근
population[['부산','광주']]
```

```python
# 불리언 인덱싱 -- 조건을 작성
population[population>2000000]
```

```python
# 시리즈 - 슬라이싱
# 인덱싱 - 인덱스 번호, 이름으로 접근하는 방법
# 인덱스 번호로 슬라이싱
display(population[1:3]) # 끝 수는 포함 안 됨 (미만)


display(population[:2])

display(population[2:])

display(population[:])
```

```python
# 인덱스 이름으로 접근
display(population['서울':'광주']) # 이름으로 접근시 미만이 아님! (포함해서 출력)
display(population[:'광주'])
display(population['부산':])
display(population[:])
```

# loc, iloc 인덱서

- 데이터 프레임의 행, 열에 접근을 더 편리하게 하기 위해 등장
- loc, iloc
- loc : 컬럼이름(라벨)명을 통해 데이터 접근하는 방법
- iloc : 컬럼의 인덱스접근을 통해 데이터 접근

- .loc(or iloc)[행,열]
  - 결국 인덱싱 슬라이싱의 규칙을 그대로 적용

```python
# loc[행, 열] 인덱스 값으로 접근
```

```python
person_info.loc['son']
```

```python
person_info.loc['son','나이']
```

```python
# 여러값 접근
person_info.loc['son',['키','나이']]
```

```python
# 여러값 접근2
person_info.loc['son',::2]
```

```python
# iloc - 인덱스 번호로 접근
person_info.iloc[0] # == person_info['son']
```

```python
#person_info.loc['son','나이']
person_info.iloc[0,2]
```

```python
# 여러값 접근
# person_info.loc['son',['키','나이']]
person_info.iloc[0, [0,2]]
```

```python
# 여러값 접근2
# person_info.loc['son',::2]
person_info.iloc[0,::2]
```

# 불리언 인덱싱

```python
person_info[person_info['키']>180]
```

```python
# .qurey() -- 불리언 인덱싱을 도와주는 함수
# 매개변수에 조건을 문자열로 작성
person_info.query('키 > 180')
```

```python
#loc 인덱서 사용해서 불리언 인덱싱
person_info.loc[person_info['키']>180]
```
