## apply()라는 함수
- np.where의 조건식 복잡해지는 것을 간단하게 작성 할 수 있도록 하는 함수


```python
# apply() 사용법

# 매개변수가 함수
# 함수를 만들어서 함수명만 작성 해주시면 됩니다!

def makeGrade(df): # 함수 실행시 조건이 될 시리즈가 필요! 매개변수로 df 작성
    grade = df['평균']

    if grade >= 25:
      return "A등급"
    elif grade >= 20:
      return "B등급"
    elif grade >= 15:
      return "C등급"
    else:
      return "F등급"
```

```python
# .apply()
mpg_copy['grade2'] = mpg_copy.apply(makeGrade,axis=1)
mpg_copy
```

```python
mpg_copy['grade2'].value_counts().sort_index()
```

  - 값이 없다

    - 수집과정에서 발생한 오류 때문에 포함 될 경우가 종종 발생

# 데이터 접근 : 인덱싱/ 슬라이싱


### 1. 인덱스 번호를 기준으로 접근
-  [  ] 로 접근
- 슬라이싱의 경우에 기준으로 시작 범위와 끝 범위를 지정
- 행, 열을 데이터 접근 시 , (comma)가 기준, 앞이 행, 뒤가 열
- loc, iloc 인덱서 존재
- loc, iloc의 경우 데이터 접근을 보다 쉽게 도와주는 역할
  - loc 인덱스 값으로 접근, iloc 인덱스 번호로 접근이 가능

### 2. 조건을 기준으로 접근 (불리언 인덱싱)
 - 조건에 따라 값을 가져오는 불리언인덱싱이 존재
 - 불리언 인덱싱 하는 방법 1
  - [  ] 를 한번 더 감싼 후 조건을 작성 예) mpg_copy.[mpg_copy['도시'] > 15 ]
- 불리언 인덱싱 하는 방법 2
  - .query() 함수 사용 :  query함수 내 매개변수에 조건을 문자열로 작성
    - 예) mpg_copy,query('mpg_copy[도시]>15')

#### (예제) category가 suv인 차들의 제조사를 묶고 해당 제조사들의 도시연비를 구한 후 barplot으로 그려보시오

```python
# 1. 카테고리 suv인 차 가지고 오기
mpg_copy[mpg_copy['category'] == 'suv']

#1.2 query()함수 사용
mpg_suv = mpg_copy.query('category == "suv"' )
mpg_suv
```

```python
# 4. barplot 그려보기
sns.barplot(data=mpg_suv_city, x='제조사', y='city_mean')
```
