## apply() 
- np.where          


```python
# apply() 

#  
#      !

def makeGrade(df): #      !  df 
    grade = df['']

    if grade >= 25:
      return "A"
    elif grade >= 20:
      return "B"
    elif grade >= 15:
      return "C"
    else:
      return "F"
```

```python
# .apply()
mpg_copy['grade2'] = mpg_copy.apply(makeGrade,axis=1)
mpg_copy
```

```python
mpg_copy['grade2'].value_counts().sort_index()
```
