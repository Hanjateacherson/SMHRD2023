```python
# requests 불러오기
# requests = 브라우저의 역할을 대신 해줌(사이트의 요청과 응답 정보를 처리)
import requests as req
```

```python
# 1. 네이버 url을 통해서 사이트 요청
# req.get("url") : 해당 url로 요청을 보내주고 나서, 응답을 받음!
# Response 200 : 통신에 성공 → 응답을 잘 받았다!
# Response 400 : 클라이언트 오류 → 우리의 실수!
# Response 500 : 서버 오류 → 우리의 잘못은 없다!
res = req.get("https://news.naver.com/")
res
# HTTP 상태를 확인할 수 있는 웹사이트
# https://developer.mozilla.org/ko/docs/Web/HTTP/Status
```

```python
# 응답 객체 전부를 받아왔다!
# 객체 안에 있는 글자 정보를 가져옴...
res.text
```

### beautifulsoup 라이브러리
- 응답받은 글자 정보를 태그별로 파싱해줌!
- 파싱 : 태그별로 쪼개준다!!

```python
from bs4 import BeautifulSoup as bs
```

```python
# 받아온 응답 글자 정보들을 html 정보로 바꿔줘!
# bs(응답받은 문자열 데이터, "파싱 방법")

soup = bs(res.text , "lxml")
soup
```

```python
# 상단의 카테고리 이름들을 가져와보자!
# html 선택자를 활용해서 지정이 가능!
# select(선택자 경로) : 리스트 형으로 반환
# select_one(선택자 경로) : 하나만 반환
top_title = soup.select('ul > li > a > span')
```

```python
soup.select_one('ul > li > a > span')
```

```python
# 태그 안에 content를 지정하기 위해서는 태그명을 제거하는 과정
# → 인덱싱 + text 속성으로 작업

news_item = []
for i in top_title :
    news_item.append(i.text)
news_item
```

```python
# 리스트 컴프리헨션 : append()는 정말 많이 쓰이는 문법이다 보니,
# 파이썬에서는 이를 한 줄로 작성하게끔 문법을 제공

news_item2 = [i.text for i in top_title]
news_item2
```

### 실습! 뉴스 타이틀을 수집해보자!
- 네이버에서 <b>눈</b> 검색 후 나오는 <b>뉴스 타이틀(제목)</b>을 수집해보자!

```python
# 1. 서버에 url 값으로 정보를 넘겨 주고, 응답 받기
res = req.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%88%88")
res.status_code # 상태 확인
```

```python
# 2. 응답받은 데이터 정보를 BeatifulSoup으로 파싱하기
soup = bs(res.text, 'lxml')
```

```python
# 3. select를 활용해서 class 또는 태그명으로 접근
snow_news_title = soup.select("div.news_contents > a.news_tit")
snow_news_title
# 요소 위에 우클릭 → copy → copy selector
# #sp_nws6 > div.news_wrap.api_ani_send > div > div.news_contents > a.news_tit
# 모든 html 요소들은 각자 본인만에 고유한 html 경로를 가지고 있음
# 이를 응용해서, 웹페이지의 구조를 살펴보고 내가 가져오고자 하는 선택자를 작성 가능!
```

```python
for i in snow_news_title :
    print(i.text)
```

```python
# 4. for문을 활용하여 list에 태그명 제거 후 append해서 출력
snow_title_content = [i.text for i in snow_news_title]
```
