```python
# Count the occurrences of each genre for each year
Genre_by_year = k_drama.groupby('Year of release')['Genre'].value_counts().unstack(fill_value=0)

# Identify the most frequent genre for each year
most_frequent_genre = Genre_by_year.idxmax(axis=1)

# Plotting a bar plot with annotations for the most frequent genre
ax = Genre_by_year.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))

# Annotate the most frequent genre for each year
for year, genre in most_frequent_genre.items():
    count = Genre_by_year.loc[year, genre]
    ax.annotate(f'{genre} ({count})', (year, count), textcoords="offset points", xytext=(0,5), ha='center')

plt.title('K-Drama Genre Distribution Over the Years')
plt.xlabel('Year of Release')
plt.ylabel('Number of Dramas')
plt.legend(title='Genre', bbox_to_anchor=(1, 1))

plt.show()
```

```python
k_drama = k_drama.assign(Genre=k_drama['Genre'].str.split(',')).explode('Genre')

# Count the occurrences of each genre for each year
Genre_by_year = k_drama.groupby('Year of release')['Genre'].value_counts().unstack(fill_value=0)

# Identify the most frequent genre for each year
most_frequent_genre = Genre_by_year.idxmax(axis=1)

# Create a DataFrame with yearly counts and most frequent genres
result_df = pd.DataFrame({
    'Year': most_frequent_genre.index,
    'Most Frequent Genre': most_frequent_genre.values,
    'Count': [Genre_by_year.loc[year, genre] for year, genre in most_frequent_genre.items()]
})

# Find newly introduced genres each year
new_genres_by_year = {}
for year in result_df['Year'].unique()[1:]:
    previous_year_genres = set(Genre_by_year.loc[year - 1].index)
    current_year_genres = set(Genre_by_year.loc[year].index)
    new_genres = current_year_genres - previous_year_genres
    new_genres_by_year[year] = ', '.join(new_genres)

# Add a new column to the result_df for newly introduced genres
result_df['New Genres'] = result_df['Year'].map(new_genres_by_year)

# Print or display the resulting DataFrame
print(result_df)
```
