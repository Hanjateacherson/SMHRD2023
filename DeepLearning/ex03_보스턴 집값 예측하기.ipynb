```python
from tensorflow.keras.datasets import boston_housing
```

```python
(X_train, y_train), (X_test, y_test) = boston_housing.load_data()
```

```python
X_train[:10]
```

```python
y_train
```

```python
# 딥러닝 모델 설계
from tensorflow.keras import Sequential # 모델의 토대, 뼈대
from tensorflow.keras.layers import Dense # 하나의 층
```

```python
# 모델 뼈대 구축
model = Sequential()

# 입구 = 입력층
# 사람들이 드나드는곳 = 데이터를 받아들이는 층
model.add(Dense(units = 30, input_dim = 13, activation = 'relu'))
# units = 생각할 수 있는 갯수
# input_dim = 입력 데이터의 특성 수, 한 행(가로)의 데이터 갯수 > 2차원일때
#                                    데이터 하나의 특성 수 > 다차원적인 의미
# activation = 'relu', 다음층에 정보를 전달하는 방법(sigmoid, tanh, relu, Maxout, elu)

# 그 외층 = 은닉층, 중간층
# 실제 건물의 역할 = 데이터를 분석하여 목적에 맞는 규칙을 찾는 층
model.add(Dense(units = 15, activation = 'relu'))
model.add(Dense(units = 15, activation = 'relu'))
# units : 아무 숫자나 적어도 무방
# activation : 분석 방법에 따라 다르지만 보통 relu(출력층 제외)
## sigmoid : 정보를 축소시켜서 전달
## relu : 좋다고 판단한 모든 정보를 전달

# 출구 = 출력층
# 사람들이 나가는곳 = 분석된 데이터의 결론을 도출하는 층
model.add(Dense(units = 1, activation = 'linear'))
# 분석 방법에따라 적을 값이 정해져있음
# units : 1(이진분류, 회귀) / 정답의 갯수(다중분류)
## 출력해야하는 값의 갯수
# activation : linear(회귀) / sigmoid(이진분류) / softmax(다중분류)
## 출력되는 값의 종류, 갯수에 따라 다른 값을 적어야한다
```

```python
model.summary()
```

```python
# 모델 학습 방법 설정
# compile
model.compile(
    loss = 'mean_squared_error', # 오차
    optimizer = 'adam', # 최적화 함수 : 여러번 학습할때 어떤식으로 최적화 할건지
    # metrics =  ['mean_squared_error'] # 평가 방법
)
```

```python
model.fit(X_train,y_train, epochs = 100,
          # validation_data = (X_test, y_test)
          validation_split = 0.3 # train 데이터중 30%를 validation 으로 사용
)
```

```python
model.evaluate(X_test,y_test)
```

```python
# 모델 저장
# 학습된 내용과 같이 저장
# 사용하고 싶을때마다 학습 없이 바로 예측 가능 > 시간 절약
model.save('/content/boston.h5')
```

```python
# 모델 불러오기
from tensorflow.keras.models import load_model
new_model = load_model('/content/boston.h5')
```

```python
new_model.evaluate(X_test,y_test)
```
