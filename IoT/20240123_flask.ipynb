```python
from flask import Flask, render_template, request
app = Flask(__name__)
@app.route('/')
def home():
    return render_template('home.html')

@app.route('/add', methods = ["GET","POST"])
def add_num():
    if request.method == "GET":
        number1 = request.args['num1']
        number2 = request.args['num2']
        result = int(number1)+int(number2)
    else:
        a = request.form['num1']
        b = request.form['num2']
        result = int(a)+int(b)
    return f'두 수의 합은 {result}입니다.'

if __name__ == "__main__":
    app.run(host = "61.80.80.214", port = 5055)
```
